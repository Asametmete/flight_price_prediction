```python
#This code divides the dataset into frequency ranges based on the "duration" values, 
#calculates the mean, sum, and count for each range, and stores the results in a new DataFrame.
```

```python
df_duration = df.groupby(["duration"]).mean("price")
```

```python
df_duration
```

```python
#drop unnecessary columns
df_duration = df_duration.drop(["Unnamed: 0","days_left"],axis=1)
```

```python
#get the durations for a column. in the above dataframe duration is an index
unique_duration = df["duration"].unique()
sorted_duration = np.sort(unique_duration)
df_duration["duration"] = sorted_duration
```

```python
#reset index and rename columns for a cleaner dataframe
df_duration = df_duration.reset_index(drop=True)
df_duration.columns= ["mean","duration"]
```

```python
df_duration
```

```python
df_duration_freq = pd.DataFrame()
df_duration_freq["frequency range"] = ""
df_duration_freq["mean"] = ""
df_duration_freq["sum"] = ""
df_duration_freq["n"] = ""
```

```python
def check_range(Duration, Temp_value):
    #check the frequency range of durations 
    Temp_value += 1 
    while not (Duration < Temp_value and Duration >= round(Temp_value - 1, 2)):
        Temp_value += 1
    return Temp_value
```

```python
# Initial frequency range start value
temp_value = 1.83  
n = 0  # Count of elements in the current frequency range
sum_value = 0  # Sum of the 'mean' values in the current range
# Create an empty DataFrame to store frequency range details
df_duration_freq = pd.DataFrame(columns=["frequency range", "mean", "n", "sum"])  

# Loop through the duration and mean columns of the DataFrame
for i, (duration, mean_value) in enumerate(zip(df_duration["duration"], df_duration["mean"])):
    if duration < temp_value and duration >= round(temp_value - 1, 2):
        # Add the mean value to the sum and increment the element count
        sum_value += mean_value  
        n += 1
    else:
        # Save the details for the current frequency range if it contains elements
        if n > 0:
            average = sum_value / n
            df_duration_freq = pd.concat(
                [df_duration_freq, pd.DataFrame({
                    "frequency range": [f"{round(temp_value-1, 2)} ile {temp_value}"],
                    "mean": [average],
                    "n": [n],
                    "sum": [sum_value]
                })],
                ignore_index=True
            )
        # Move to the next frequency range
        temp_value = check_range(duration, temp_value)
        sum_value = mean_value  # Start a new sum with the current mean value
        n = 1  # Reset the element count for the new range

# Process the remaining range after exiting the loop
if n > 0:
    average = sum_value / n
    df_duration_freq = pd.concat(
        [df_duration_freq, pd.DataFrame({
            "frequency range": [f"{round(temp_value-1, 2)} ile {temp_value}"],
            "mean": [average],
            "n": [n],
            "sum": [sum_value]
        })],
        ignore_index=True
    )
```

<h1>Encoding and Normalization for the Training Model<h1>

```python
df.drop(columns=["destination_city","source_city",'Unnamed: 0','departure_time','arrival_time','flight'],inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse_output = False)
encoded_data = encoder.fit_transform(df[["airline"]])

#Create a dataframe because of store encoded_data
encoded_df = pd.DataFrame(encoded_data,columns = encoder.get_feature_names_out(["airline"]))

#merge the these two dataset
df = pd.concat([df,encoded_df],axis=1)
```

```python
encoder = OneHotEncoder(sparse_output = False)
encoded_data2 = encoder.fit_transform(df[["stops"]])

#Create a dataframe because of store encoded_data
encoded_df2 = pd.DataFrame(encoded_data2,columns = encoder.get_feature_names_out(["stops"]))

#merge the these two dataset
df = pd.concat([df,encoded_df2],axis=1)
```

```python
df.drop(columns=["airline","stops",'class'],inplace=True)
```

```python
#set the these there columns between 0 and 1
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
df["duration"] = scaler.fit_transform(df[["duration"]])
df["days_left"] = scaler.fit_transform(df[["days_left"]])
df["price"] = scaler.fit_transform(df[["price"]])
```

<h1>Model Training</h1>

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("price",axis=1)
y = df["price"]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
# Create the Sequential model
model = Sequential()

# Add the input layer and the first hidden layer
model.add(Dense(64, input_shape=(x_train.shape[1],), activation="relu"))  # 64 neurons, ReLU activation
model.add(Dropout(0.3))  # Dropout to prevent overfitting

# Add the second hidden layer
model.add(Dense(32, activation="relu"))  # 32 neurons, ReLU activation
model.add(Dropout(0.3))

# Add the output layer
model.add(Dense(1, activation="linear"))  # Single neuron for regression, linear activation

# Compile the model
model.compile(optimizer="adam", loss="mse", metrics=["mae"])  # Mean Squared Error (MSE) as loss, Mean Absolute Error (MAE) as metric

# Train the model
history = model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),  # Validate on the test set
    epochs=50,  # Train for 50 epochs
    batch_size=32,  # Batch size of 32
    verbose=1  # Display training progress
)
```

```python
# Evaluate the model on the test set
results = model.evaluate(x_test, y_test)
print("Test Loss (MSE):", results[0])
print("Test MAE:", results[1])
```

```python
# Predict prices using the trained model
y_pred = model.predict(x_test)
```

```python
#Display actual vs predicted prices for the first few test examples
for i in range(10):
    print(f"Actual: {y_test.iloc[i]}, Predicted: {y_pred[i][0]}")
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
test_r2 = r2_score(y_test, y_pred)
print(f"Test R² Score: {test_r2:.2f}")
```
